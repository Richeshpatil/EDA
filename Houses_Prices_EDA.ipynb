# Categorical Columns

```python
total = categorical_columns.isnull().sum().sort_values(ascending=False)
percent = (categorical_columns.isnull().sum()/categorical_columns.isnull().count()*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total Missing Count', ' % of Total Observations'])
missing_data.index.name ='Categorical_Columns'
missing_data.head(20)
```

```python
missing_values = categorical_columns.isnull().sum(axis=0).reset_index()
missing_values.columns = ['column_name', 'missing_count']
missing_values = missing_values.loc[missing_values['missing_count']>0]
missing_values = missing_values.sort_values(by='missing_count')

ind = np.arange(missing_values.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(12,7))
rects = ax.barh(ind, missing_values.missing_count.values, color='red')
ax.set_yticks(ind)
ax.set_yticklabels(missing_values.column_name.values, rotation='horizontal')
ax.set_xlabel("Missing Observations Count")
ax.set_title("Missing Observations Count - Categorical Features")
plt.show()
```

On the Observation Some Categorical Features Has the Most of Missing Values on the obsevation of Charts And Pivot Table. for the Modelling Purpose we Have to Drop some Columns that are Greater Than 50%
